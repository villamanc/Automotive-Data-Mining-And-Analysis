## Clean The Data
- Handle missing values
- Remove Uneccesary Columns
- Transform the data for analysis

```python
df
```

```python
#Replace Na values in the 'price_drop' column with 0 (if there is no price drop then the price drop is 0)
df['price_drop'] = df['price_drop'].fillna(0)
```

```python
#Drop the 'driver_rating', and 'driver_reviews_num' columns (column isn't important to our analysis)
df = df.drop(columns=['driver_rating'])
df = df.drop(columns=['driver_reviews_num'])
```

```python
df.isna().sum()
```

```python
#Remove rows where 'seller_name' is Na (remove unknown sellers)
df = df[df['seller_name'].notna()]

#Fill Na values with 0 for seller_rating
df['seller_rating'] = df['seller_rating'].fillna(0)
```

```python
#Set any Electric Cars to have an MPG of ('No Fuel Used')
df.loc[df['fuel_type'] == 'Electric', 'mpg'] = 'No Fuel Used'
```

```python
#Remove rows where 'fuel_type' is Na
df = df[df['fuel_type'].notna()]
```

```python
df.isna().sum()
```

```python
#Drop remaining rows containing Na
df = df.dropna()
```

```python
df.dtypes
```

```python
df
```

```python
#Remove any car above 2 mill
df = df[df['price'] <= 2000000]
#Remove any car below 5000
df = df[df['price'] >= 5000] 

df
```

```python
#Extract the max MPG from the mpg column, and set the data type as a float
def extract_mpg (value):
    if '-' in value:
        return value.split('-')[1]
    elif 'No Fuel Used' in value:
        return 0
    else:
        return value

    
#Apply the function to the 'mpg' column and convert the result to integer
df['max_mpg'] = df['mpg'].apply(extract_mpg).astype(float)

df
```

```python
df.dtypes
```

```python
#change the year data type to an object
df['year'] = df['year'].astype(str)
```

```python
#reorder the columns
df.columns

columns_reordered = ['manufacturer', 'model', 'year', 'mileage', 'engine', 'transmission',
       'drivetrain', 'fuel_type', 'mpg','max_mpg', 'exterior_color', 'interior_color',
       'accidents_or_damage', 'one_owner', 'personal_use_only', 'seller_name',
       'seller_rating', 'price_drop', 'price']

df = df[columns_reordered]
df
```
